# Activit√© 1

```python
import matplotlib.pyplot as plt
import numpy as np

def calculate_global_cost(cout_variable, cout_unitaire, nb_produits):
    cout_global = cout_variable + cout_unitaire * nb_produits
    return cout_global


cout_variable = 100
cout_unitaire = 10
nb_produits = np.arange(0, 101, 1)  

# Tracer le co√ªt global en fonction du nombre de produits pour diff√©rents co√ªts unitaires
plt.figure(figsize=(10, 6))
for cout_unitaire in [5, 10, 15]:
    couts_globaux = [calculate_global_cost(cout_variable, cout_unitaire, nb_produits) for nb_produits in nb_produits]
    plt.plot(nb_produits, couts_globaux, label=f'cout_unitaire={cout_unitaire}')

plt.xlabel('Nombre de produits')
plt.ylabel('Co√ªt global')
plt.title('Co√ªt global en fonction du nombre de produits pour diff√©rents co√ªts unitaires')
plt.legend()
plt.grid(True)
plt.show()

# Tracer le co√ªt global en fonction du nombre de produits pour diff√©rents co√ªts variables
plt.figure(figsize=(10, 6))
for cout_variable in [50, 100, 150]:
    couts_globaux = [calculate_global_cost(cout_variable, cout_unitaire, nb_produits) for nb_produits in nb_produits]
    plt.plot(nb_produits, couts_globaux, label=f'cout_variable={cout_variable}')

plt.xlabel('Nombre de produits')
plt.ylabel('Co√ªt global')
plt.title('Co√ªt global en fonction du nombre de produits pour diff√©rents co√ªts variables')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

matrice = np.array([[10, 5], [-4, 1]])
x = 9000
y = 50

inverse = np.linalg.inv(matrice)
print("Inverse de la matrice:")
print(inverse)
vecteur = np.array([x, y])
resultat = np.dot(inverse, vecteur)
print("Produit de la matrice et du vecteur:")
print(resultat)
```

Et meme si la multiplication √©tait possible, le r√©sultat sera different pour les deux produits.

```python
#3
import math

def round_down(x): 
    return math.floor(x) # floor arrondi √† la moyenne inf

def round_up(y):
    return math.ceil(y*100)/100


# Example usage
x = 5.7
y = 3.456

x_rounded = round_down(x)
y_rounded = round_up(y)

print("x =", x_rounded)
print("y =", y_rounded)
```

# Activit√© 3

```python
import numpy as np

def generate_random_problem(size):
    
    matrice = np.random.randint(-10, 10, size=(size, size))
    vector = np.random.randint(-10, 10, size=(size,))
    return matrice, vector

def has_solution(matrice):
    """
    V√©rifier si la matrice a une solution unique en calculant son d√©terminant.
    """
    det = np.linalg.det(matrice)
    return det != 0

def solve(matrice, vector):
    """
    R√©sout le syst√®me d'√©quations si une solution unique existe.
    """
    if has_solution(matrice):
        inverse = np.linalg.inv(matrice)
        return np.dot(inverse, vector)
    else:
        return None

def main():
    N = 1000
    size = 5
    counter = 0

    for _ in range(N):
        matrice, vector = generate_random_problem(size)
        if has_solution(matrice):
            counter += 1
            solution = solve(matrice, vector)

    percentage_solvable = (counter / N) * 100
    percentage_not_solvable = round(100 - percentage_solvable, 2)

    print(f"Pourcentage de probl√®mes solvables: {percentage_solvable}%")
    print(f"Pourcentage de probl√®mes non solvables: {percentage_not_solvable}%")

if __name__ == "__main__":
    main()
```

3- Un syst√®me est ind√©termin√© quand il a une infinit√© de solutions (le det=0)
Un syst√®me est incompatible quand il n'a pas de solution

```python
import numpy as np

def define_system():
    # Matrice des coefficients A
    A = np.array([
        [1, 2, -1, 1, 1],
        [2, -1, 3, 0, 4],
        [-1, 3, 2, 1, -1],
        [3, -1, 1, 2, 3],
        [2, 1, -1, -3, 1]
    ])
    
    # Vecteur des r√©sultats Y
    Y = np.array([8, 15, 7, 20, 5])
    
    return A, Y

# ON utilise les memes fonctions que l'activit√© 3 

def has_solution(matrice):
    
    det = np.linalg.det(matrice)
    return det != 0

def solve(matrice, vector):
  
    if has_solution(matrice):

        return np.linalg.solve(matrice, vector)
    else:
        return None

def main():
    # D√©finir le syst√®me d'√©quations
    A, Y = define_system()

    print("A :")
    print(A)
    print("Y :")
    print(Y)
    
    # V√©rifier si une solution existe(det n'est pas nul)
    if has_solution(A):
        print("\nLe syst√®me poss√®de une solution unique.")
        solution = solve(A, Y)
        print("\nSolution :")
        print(solution)
    else:
        print("\nLe syst√®me ne poss√®de pas de solution unique.")

if __name__ == "__main__":
    main()
```
